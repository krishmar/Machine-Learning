# 2. Data Manuplation

a. Create new column 'Total_benefitted' which is a sum of inmates benefitted through all modes  

```python
total_benefitted = df['No. of Inmates benefitted by Elementary Education'] + df['No. of Inmates benefitted by Adult Education'] + df['No. of Inmates benefitted by Higher Education'] + df['No. of Inmates benefitted by Computer Course']
df['Total_benefitted'] = total_benefitted
df
```

b. Create new row 'Total_benefitted' which is a is the sum of all inmates benefitted through each mode across all states.

```python
new_row = df.sum(axis=0,numeric_only=True)
df = df.append(new_row, ignore_index=True)
df[['STATE/UT','No. of Inmates benefitted by Elementary Education','No. of Inmates benefitted by Adult Education','No. of Inmates benefitted by Higher Education','No. of Inmates benefitted by Computer Course','Total_benefitted']]
```

# 2. Data Plotting:

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

a. Bar chart to show which state has the maximum number of beneficiaries

```python
plt.figure(figsize=(15,4))
ax=sns.barplot(y='Total_benefitted',x='STATE/UT',data=df)
ax.set_xticklabels(ax.get_xticklabels(), rotation=40, ha="right")
plt.show()
```

b. Pie chart to depict the ratio among different modes of benefits

```python
fig1, ax1 = plt.subplots()
data_in_use = df[['No. of Inmates benefitted by Elementary Education','No. of Inmates benefitted by Adult Education','No. of Inmates benefitted by Higher Education','No. of Inmates benefitted by Computer Course']]
sizes = data_in_use.iloc[35,:]
labels = data_in_use.columns
ax1.pie(sizes, labels=labels, autopct='%1.1f%%',shadow=True, startangle=90)
ax1.axis('equal')
plt.show()
```
