## Classfication by K-means cluster

```python
# Importing libraries necessary for this project
import numpy as np
import pandas as pd
import seaborn as sns

import matplotlib.pyplot as plt
%matplotlib inline

from PIL import Image
import numpy

from sklearn.cluster import KMeans
```

#### 1. Open and display the image “dog.jpeg”.image “dog.jpeg”.

```python
im = Image.open("dogs.jpeg")
plt.figure(figsize = (15,8))
plt.imshow(im)
```

#### 2.Convert image in to a two-dimensional array.

```python
np_im_3D = numpy.array(im)
np_im = np_im_3D.reshape(-1,2)
#print(np_im.shape)
df=pd.DataFrame(np_im)
df.columns = ['col_1','col_2']
```

#### 3. Cluster image to K-means cluster

```python
kmeans = KMeans(n_clusters=3)
```

```python
kmeans.fit(df)
```

```python
x, y, z = np_im_3D.shape
np_im_3D.shape
```

```python
cluster_centers = kmeans.cluster_centers_
cluster_centers
```

#### 4. cluster label of every pixel in the image

```python
cluster_labels = kmeans.labels_
cluster_labels
```

#### Plot it back as an image

```python
plt.figure(figsize = (15,8))
out = cluster_centers[cluster_labels].reshape(x, y, z)
#plt.imshow(cluster_centers[cluster_labels].reshape(x, y, z))
plt.imshow((out * 255).astype(np.uint8))
```

```python
df['cluster'] = kmeans.labels_
sns.set_style('whitegrid')
sns.lmplot('col_1', 'col_2',data=df, hue='cluster',
           palette='coolwarm',size=6,aspect=1,fit_reg=False)
```
